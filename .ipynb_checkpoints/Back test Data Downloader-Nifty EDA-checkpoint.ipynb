```python
# Import modules
from nsepy import get_history
from datetime import datetime, timedelta
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like #For solving import pandas_datareader issue
import numpy as np
import datetime
import csv
import requests
import pandas_datareader.data as web
import pandas_datareader as pdr
from pandas_datareader import data, wb
from datetime import date
#to plot within notebook
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = r"A:\FinMinII\Trading\NSE_Indices\Backtest\Nifty_Backtest"
file = "Backtest RSI and Nifty.csv"#


# # # Input Start and End Date
start = datetime.datetime(2011,1,7) # keep the strt date starting from Friday for weekly analysis (Friday is first day for expirty week and month)
end = datetime.datetime(2022,2,2)
```

```python
import os
os.getcwd()
os.chdir(path)
os.getcwd()
```

```python
NIFTY_opt = get_history(symbol="NIFTY",
                        start=start,
                        end=end,
                        index=True)

```

```python
NIFTY_opt = NIFTY_opt.reset_index(level=['Date'])
NIFTY_opt['Date'] = pd.to_datetime(NIFTY_opt['Date'])


```

```python
# Creating Historical date data frame to handle holidays

DS = pd.date_range(start=start, end=end)
Nifty_Historical_DF = pd.DataFrame({'Date': DS})
```

```python
# Merging Nifty data with Date DataFrame

NIFTY_opt = pd.merge(Nifty_Historical_DF,NIFTY_opt,on = "Date", how = 'left' )


NIFTY_opt['Day_of_Week'] = NIFTY_opt['Date'].dt.day_name()

NIFTY_opt['week']=NIFTY_opt.Date.dt.week
```

```python
NIFTY_opt = NIFTY_opt[NIFTY_opt.Day_of_Week != 'Saturday'] 
NIFTY_opt = NIFTY_opt[NIFTY_opt.Day_of_Week != 'Sunday']
```

```python
# NIFTY_opt.diff
NIFTY_opt['Weekly Close'] = NIFTY_opt['Close'] - NIFTY_opt['Open'].shift(4)
NIFTY_opt['Weekly %Chng'] = (NIFTY_opt['Weekly Close']/NIFTY_opt['Open'].shift(4))
```

```python
Nifty_Weekly_Close = NIFTY_opt[NIFTY_opt.Day_of_Week == 'Thursday'] 
```
